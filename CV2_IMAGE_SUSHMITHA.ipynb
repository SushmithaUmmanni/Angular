```
import cv2
from google.colab.patches import cv2_imshow
```

```
# Load an color image in grayscale
img = cv2.imread('indiragandhi.jpg')
print(img)
```

display an image

```

cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

ROTATE AN IMAGE USING OPENCV

```
rotate_img1=cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotate_img1)
```

```
rotate_img2=cv2.rotate(img,cv2.ROTATE_180)
cv2_imshow(rotate_img2)

rotate_img3=cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
cv2_imshow(rotate_img3)
```

```
resize_img=cv2.resize(img,(250,250))
cv2_imshow(resize_img)
```

RESIZING AN IMAGE USING INTERPOLATION

```
resize_img_interpolation=cv2.resize(img,(300,300),interpolation=cv2.INTER_LINEAR)
cv2_imshow(resize_img_interpolation)
```

Simple Image Thresholding

```
#Thresholding can only be applied to grayscale images.
gray_img=cv2.imread("indiragandhi.jpg",0)
ret,treshold_binary=cv2.threshold(gray_img,25,325,cv2.THRESH_BINARY)
cv2_imshow(treshold_binary)
#cv2_imshow(int(ret))
```

```
ret,treshold_binary=cv2.threshold(gray_img,25,325,cv2.THRESH_BINARY_INV)
cv2_imshow(treshold_binary)
```

```
ret,treshold_binary=cv2.threshold(gray_img,25,35,cv2.THRESH_MASK)
cv2_imshow(treshold_binary)
```

```
et,treshold_binary=cv2.threshold(gray_img,25,325,cv2.THRESH_TRIANGLE)
cv2_imshow(treshold_binary)
```
